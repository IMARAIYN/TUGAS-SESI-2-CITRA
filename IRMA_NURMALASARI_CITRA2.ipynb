```python
import numpy as np
import imageio
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan channel warna dan hasil konversi
def display_channels_and_conversions(image_path):
    # Membaca gambar
    image = imageio.v3.imread(image_path)

    # Memisahkan Channel RGB
    R = image[:, :, 0]
    G = image[:, :, 1]
    B = image[:, :, 2]

    # Konversi ke Grayscale
    grayscale = np.dot(image[..., :3], [0.299, 0.587, 0.114])

    # Konversi ke Threshold/Biner (grayscale threshold, nilai threshold 127)
    threshold = np.where(grayscale > 127, 255, 0)

    # Menampilkan hasil
    fig, ax = plt.subplots(2, 3, figsize=(12, 8))

    # Menampilkan gambar asli
    ax[0, 0].imshow(image)
    ax[0, 0].set_title("Original Image")

    # Menampilkan Channel R
    ax[0, 1].imshow(R, cmap="Reds")
    ax[0, 1].set_title("Channel R (Red)")

    # Menampilkan Channel G
    ax[0, 2].imshow(G, cmap="Greens")
    ax[0, 2].set_title("Channel G (Green)")

    # Menampilkan Channel B
    ax[1, 0].imshow(B, cmap="Blues")
    ax[1, 0].set_title("Channel B (Blue)")

    # Menampilkan gambar Grayscale
    ax[1, 1].imshow(grayscale, cmap="gray")
    ax[1, 1].set_title("Grayscale Image")

    # Menampilkan gambar Threshold/Biner
    ax[1, 2].imshow(threshold, cmap="gray")
    ax[1, 2].set_title("Threshold Image (Binary)")

    # Menghilangkan sumbu
    for a in ax.ravel():
        a.axis("off")

    plt.tight_layout()
    plt.show()

# Masukkan path gambar daun di sini
display_channels_and_conversions("/content/sample_data/daun/daun pepaya.jpg")
display_channels_and_conversions("/content/sample_data/daun/daun singkong.jpg")
display_channels_and_conversions("/content/sample_data/daun/daun keninir.jpg")
```
